```python
import matplotlib.pyplot as plt
from random import randint as rad

X = list(range(100))

Y = []

def Y_random():
    Y = []
    for i in X:
        Y.append(rad(0,100))
        
    return Y
    
for i in range(10):
    plt.plot(X,Y_random(),label = "Plot number" + str(i+1))
    
plt.rc("figure",figsize = (20,6))
# plt.plot(X,Y,'ko--')

plt.legend(loc = "lower left")
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title("Random figure")
plt.xlim(0,100)
plt.ylim(0,101)
plt.show()
```
